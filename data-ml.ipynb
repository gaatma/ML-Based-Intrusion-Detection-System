## Data Exploration

```python
#DATA & EXPLORATION
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Column names for the NSL-KDD dataset

columns = [
    'duration', 'protocol_type', 'service', 'flag', 'src_bytes',
    'dst_bytes', 'land', 'wrong_fragment', 'urgent', 'hot',
    'num_failed_logins', 'logged_in', 'num_compromised', 'root_shell',
    'su_attempted', 'num_root', 'num_file_creations', 'num_shells',
    'num_access_files', 'num_outbound_cmds', 'is_host_login',
    'is_guest_login', 'count', 'srv_count', 'serror_rate',
    'srv_serror_rate', 'rerror_rate', 'srv_rerror_rate',
    'same_srv_rate', 'diff_srv_rate', 'srv_diff_host_rate',
    'dst_host_count', 'dst_host_srv_count', 'dst_host_same_srv_rate',
    'dst_host_diff_srv_rate', 'dst_host_same_src_port_rate',
    'dst_host_srv_diff_host_rate', 'dst_host_serror_rate',
    'dst_host_srv_serror_rate', 'dst_host_rerror_rate',
    'dst_host_srv_rerror_rate', 'label', 'difficulty'
]

#Load the dataset
url_train = "https://raw.githubusercontent.com/defcom17/NSL_KDD/master/KDDTrain+.txt"
df = pd.read_csv(url_train, names=columns)

print("Dataset loaded")
print(f"Shape: {df.shape}") #property that returns two numbers: (Number of Rows, Number of Columns).
```

```python
#To see the first 5 rows
df.head()
```

```python
#What type of data do we have 

df.dtypes
```

```python
#any missing values?
print(df.isnull().sum().sum())
```

```python
#Understanding the labels
#what attack type exist

print(df['label'].value_counts())
```

```python
#simplify normal verses attack (binary classification)

df['label_binary'] = df['label'].apply(lambda x: 'normal' if x == 'normal' else 'attack')
```

```python
#How balanced are the classess?
sns.countplot(x='label_binary', data=df,palette='Set2')
plt.title('Normal vs Attack Traffic')
plt.xlabel('Label')
plt.ylabel('Count')
plt.show()

```

```python
#Which protocols appeared most?
sns.countplot(x='protocol_type', hue='label_binary',data=df, palette='Set2')
plt.title('Protocol Type by Label')
plt.show()
```

```python
#Distribution of connection duration
plt.figure(figsize=(10,4))
sns.histplot(df[df['duration'] < 100]['duration'], bins=50, kde=True)
plt.title('Distribution of Connection Duration (< 100s)')
plt.show()
```
